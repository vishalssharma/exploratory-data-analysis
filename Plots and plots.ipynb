## Seaborn
The python visualization libraby Seaborn is based on matplotlib and provides a high-level interface for drawing attractive statistical graphics.

### The basic steps are 
* 1) Prepare some data 
* 2) Control figure aesthetics 
* 3) Plot with Seaborn 
* 4) Further customize your plot

```python
sns.set(color_codes=True)
tips = sns.load_dataset("tips") ##step1
sns.set_style('whitegrid')  ##step2
g = sns.lmplot(y="total_bill", x="tip",hue='smoker', data=tips,aspect=2)  ##step3
g = (g.set_axis_labels("Tip","Total bill(USD)").set(xlim=(0,10),ylim=(0,100))) ##step4
plt.title('title')
plt.show(g)
```
